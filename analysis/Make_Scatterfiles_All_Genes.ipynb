```python
uniprotmap = uniprotmap.loc[uniprotmap['struct'].isin(clumps_results['PDBID-CHAIN'])]
uniprotmap = uniprotmap.loc[uniprotmap['u1'].isin(clumps_results['UNIPROT_ID'])]
uniprotmap.reset_index()
uniprotmap
```

```python
import os

out_dir = 'uniprot_chunks'
chunk_length= 300
if not os.path.exists(out_dir):
    os.mkdir(out_dir)
gl = len(uniprotmap.index)
start = 0
end = chunk_length
i = 0
while end < gl:
    if i < 10:
        num = '0000%s' %i
    elif i < 100:
        num = '000%s' %i
    elif i < 1000:
        num = '00%s' %i
    fn = 'uniprot_chunks/allgenesstructmapped_%s.gz' %num
    uniprotmap.iloc[start:end].to_csv(fn,
            sep='\t',
            header=False,
            index=False)
    i = i + 1
    start = end
    end = end+chunk_length
end = gl

if i < 10:
    num = '0000%s' %i
elif i < 100:
    num = '000%s' %i
elif i < 1000:
    num = '00%s' %i
fn = 'uniprot_chunks/allgenesstructmapped_%s.gz' %num
uniprotmap.iloc[start:end].to_csv(fn,
        sep='\t',
        header=False,
        index=False)
```
